```python
import pandas as pd
import numpy as np
```

```python
final = pd.read_csv('NOAA Temp/14_01.csv')
final = final.drop(columns = '3')
dti = pd.date_range('01/01/2014', periods=1826, freq='D')
for j in range(14,19):
    for i in range(1,13):
        if j == 14 and i == 1: continue
        temp = pd.read_csv('NOAA Temp/'+str(j)+'_{:02d}'.format(i)+'.csv')
        temp = temp.drop(temp.columns[[0]], axis=1)
        final = pd.concat([final, temp], sort=False)
final['Date'] = dti
```

```python
for j in ['Ozone', 'CO']:
    final2 = pd.read_csv(j+'/datafile2014.txt', sep=',', nrows=31)
    final2 = final2.drop(columns = 'dom')
    temp = final2.values
    temp = temp.astype(np.float)
    temp = temp.flatten('F')
    temp = temp[np.isfinite(temp)]
    for i in range(2015,2019):
        final2 = pd.read_csv(j+'/datafile'+str(i)+'.txt', sep=',', nrows=31)
        final2 = final2.drop(columns = 'dom')
        temp2 = final2.values
        temp2 = temp2.astype(np.float)
        temp2 = temp2.flatten('F')
        temp2 = temp2[np.isfinite(temp2)]
        print(len(temp2))
        temp = np.append(temp, temp2)
    final[j] = temp
```

```python
final = final.reset_index(drop=True)
```

```python
for ind_i in range(len(final)):

    for ind_j, _ in enumerate(energy_consumption):
        
        if str(energy_consumption['Year'][ind_j]) in str(final['Date'][ind_i]):
            
            final.loc[ind_i, 'Electricity(GWh)/Non-Residential'] = energy_consumption['Electricity(GWh)/Non-Residential'][ind_j]
            final.loc[ind_i, 'Electricity(GWh)/Residential'] = energy_consumption['Electricity(GWh)/Residential'][ind_j]
            final.loc[ind_i, 'Gas(Therms)/Non-Residential'] = energy_consumption['Gas(Therms)/Non-Residential'][ind_j]
            final.loc[ind_i, 'Gas(Therms)/Residential'] = energy_consumption['Gas(Therms)/Residential'][ind_j]

            continue
            
    for ind_k, _ in enumerate(household_income):
        
        if ind_k > len(household_income)-1:
            break
        
        if str(household_income['Year'][ind_k]) in str(final['Date'][ind_i]):
            
            final.loc[ind_i, 'Median_Income(dollars)'] = household_income['Median_House_Income (dollars)'][ind_k]
            
            continue
            
            
    for ind_m, _ in enumerate(population):
        
        if ind_m > len(population)-1:
            break

        if str(population['Year'][ind_m]) in str(final['Date'][ind_i]):
            
            
            final.loc[ind_i, 'Population'] = population['Population'][ind_m]
            
            continue
    
    for ind_n, _ in enumerate(energy_percentages):
        if ind_n > len(energy_percentages)-1:
            break
        if str(energy_percentages['Year'][ind_n]) in str(final['Date'][ind_i]):
            
            final.loc[ind_i, 'Wind'] = energy_percentages['Wind'][ind_n]
            final.loc[ind_i, 'Solar'] = energy_percentages['Solar'][ind_n]
            final.loc[ind_i, 'Small hydro'] = energy_percentages['Small hydro'][ind_n]
            final.loc[ind_i, 'Geothermal'] = energy_percentages['Geothermal'][ind_n]
            final.loc[ind_i, 'Biomass'] = energy_percentages['Biomass'][ind_n]
            final.loc[ind_i, 'Large hydro'] = energy_percentages['Large hydro'][ind_n]
            final.loc[ind_i, 'Coal'] = energy_percentages['Coal'][ind_n]
            final.loc[ind_i, 'Nuclear'] = energy_percentages['Nuclear'][ind_n]
            final.loc[ind_i, 'Natural gas'] = energy_percentages['Natural gas'][ind_n]
            final.loc[ind_i, 'Unspecified'] = energy_percentages['Unspecified'][ind_n]
            
            continue
        
```
